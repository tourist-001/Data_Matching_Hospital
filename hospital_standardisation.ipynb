#### Mettre tous les colonne type string en majuscul

```python
h1 = h1.astype(str).apply(lambda x: x.str.upper())
h2 = h2.astype(str).apply(lambda x: x.str.upper())
h1.head(20)
```

#### Pour l'adresse, on enlève “ROUTE”,"STREET" “ST”, "AVE", "ROAD" “PLACE”, “RD”, etc.

```python
import nltk
from nltk.tokenize.treebank import TreebankWordDetokenizer

name_stopword = ["ROUTE","STREET", "ST", "AVE", "ROAD", "PLACE", "RD"] 
h1['Address_token'] = h1['Address'].apply(nltk.word_tokenize)
h1['Address_clean'] = h1['Address_token'].apply(lambda x: [word for word in x if word not in name_stopword])
h1['Address_clean'] = h1['Address_clean'].apply(TreebankWordDetokenizer().detokenize)
h2['Address_token'] = h2['Address'].apply(nltk.word_tokenize)
h2['Address_clean'] = h2['Address_token'].apply(lambda x: [word for word in x if word not in name_stopword])
h2['Address_clean'] = h2['Address_clean'].apply(TreebankWordDetokenizer().detokenize)
h2
```

#### Pour les villes, on calcule la distance levenshtein, et si le score est < 2, on considère que c'est une erreur écriture et on fait la correction.

```python
h1['City'].value_counts().sort_index()
```

### Je vais supprimer cette partie car je trouve ABINGDON et ABINGTON c'est totalement 2 villes différentes

```python
some_values = ['ABINGDON','ABINGTON']
h1.loc[h1['City'].isin(some_values)]
```

#### Pour le code postal, on garde que les chiffres, et on met tous les valeurs qui sont pas de 5 chiffres en 0. Par exemple un code postale de 3030, on le considère null.

```python
h2['ZIP Code'] = h2['ZIP Code'].astype(str)
h2['ZIP Code'] = h2['ZIP Code'].str.findall('[0-9]+')
h2['ZIP Code'] = h2['ZIP Code'].str.join("")
h2['ZIP Code'] = h2['ZIP Code'].fillna("")
h2['ZIP Code'] = h2['ZIP Code'].apply(lambda x: x if len(x)== 5 else 0)
h1['ZIP Code'] = h1['ZIP Code'].astype(str)
h1['ZIP Code'] = h1['ZIP Code'].str.findall('[0-9]+')
h1['ZIP Code'] = h1['ZIP Code'].str.join("")
h1['ZIP Code'] = h1['ZIP Code'].fillna("")
h1['ZIP Code'] = h1['ZIP Code'].apply(lambda x: x if len(x)== 5 else 0)
h1.head(20)
```
